# Introduction

Citi Bike is the nation's largest bike share program, with 25,000 bikes and over 1,500 stations across Manhattan, Brooklyn, Queens, the Bronx, Jersey City and Hoboken all in New York City, United States. People within NYC get to ride bikes provided by Citi Bike from stations that warehouse the bikes to the destination of the commuters, and after each ride, the bike will be returned back to the nearest station with NYC.

## How it works
### Join
- Become an Annual Member or buy a short-term pass through the Citi Bike app.

### Unlock 
- Find an available bike nearby, and get a ride code or use your member key to unlock it.

### Ride 
- Take as many short rides as you want while your pass or membership is active.

### Return 
- Return your bike to any station, and wait for the green light on the dock to make sure it's locked.


# Project Goal 

We want to analyse the cummulative data generated per trip for individual ride within the timeframe. At the end of this project, we would be able to provide answers to some questions that might guide the riders and the service provider in the future. Top to this project is trip_duration variable. Since duration determines the price, we want to investigate other variables in relation to trip_duration as target variable. With the enormous data at hand, for this project, we would only be focusing on rides that occured in the **YEAR 2014 `alone`**.  

## Dataset Content

This data set includes information about individual rides made in a bike-sharing system covering the New York City. In the dataset, we have 14 features (additional features will be engineered as the project progresses) and several observations (running to millions).

## Features Info 

- Trip Duration: trip in seconds
- Start Time and Date: trip start time and date
- Stop Time and Date: trip stop time and date 
- Start Station ID: trip start station unique ID
- Start Station Name: trip start station name
- Start Station Latitude: trip start station latitude
- Start Station Longitude: trip start station longitude
- End Station ID: trip end station unique ID
- End Station Name: trip end station name
- End Station Latitude: trip end station latitude
- End Station Longitude: trip end sation longitude
- BikeID: unique ID of the bikes 
- User Type: (Customer = 24-hour pass or 3-day pass user; Subscriber = Annual Member)
- Gender: (Zero=unknown; 1=male; 2=female)
- Year of Birth: birth year of the riders
